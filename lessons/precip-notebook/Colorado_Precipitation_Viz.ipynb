```python
from tempfile import NamedTemporaryFile

VIDEO_TAG = """<video controls>
 <source src="data:video/x-m4v;base64,{0}" type="video/mp4">
 Your browser does not support the video tag.
</video>"""

def anim_to_html(anim):
    if not hasattr(anim, '_encoded_video'):
        with NamedTemporaryFile(suffix='.mp4') as f:
            anim.save(f.name, fps=20, extra_args=['-c', 'libx264'],writer='avconv')
            video = open(f.name, "rb").read()
        anim._encoded_video = video.encode("base64")
    
    return VIDEO_TAG.format(anim._encoded_video)
```

```python
from IPython.display import HTML

def display_animation(anim):
    plt.close(anim._fig)
    return HTML(anim_to_html(anim))
```

```python
from matplotlib import animation

# First set up the figure, the axis, and the plot element we want to animate
fig = plt.figure()
ax = plt.axes(xlim=(-110, -101), ylim=(36, 42))
ax.scatter(p[50][:,3],p[50][:,2],c=p[50][:,-1],linewidths=0,s=30,cmap=mpl.cm.jet,alpha=0.7)

# initialization function: plot the background of each frame
def init():
    scatter = ax.scatter(p[0][:,3],p[0][:,2],c=p[0][:,-1],linewidths=0,s=30,cmap=mpl.cm.jet,alpha=0.7)
    return scatter

# animation function.  This is called sequentially
def animate(i):
    scatter = ax.scatter(p[i][:,3],p[i][:,2],c=p[i][:,-1],linewidths=0,s=30,cmap=mpl.cm.jet,alpha=0.7)
    return scatter

# call the animator.  blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=103, interval=20, blit=True)

# call our new function to display the animation
display_animation(anim)
```

```python
import matplotlib as mpl
import matplotlib.cm
import matplotlib.colors
import numpy as np

def data_to_colors(data, colormap=mpl.cm.jet):
    c = [mpl.colors.rgb2hex(d[0:3]) for d in colormap(mpl.colors.Normalize()(data))]
    return c
```

```python
circles=[]
for lat,lon,precip,c in zip(p[87][:,2],p[87][:,3],p[87][:,-1],data_to_colors(p[87][:,-1])):
    cm = Circle(location=(lat,lon), radius=2500, weight=2,
                  color=c, opacity=0.0, fill_opacity=0.5,
                  fill_color=c)
    m.add_layer(cm)
```
